```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt

data_dir = [6]

for d in data_dir:
    All_files = sorted(glob('/home/torenvln/git/bnp-anomaly/data/fail/' + str(d) + '/*.csv'))
    print(All_files)
    headers = ('x','y')
    df = pd.read_csv(All_files[1], names=headers)
    df = df.iloc[107000:125000]
    df.reset_index(drop=True, inplace=True)
    x = df.index
    y = df['y']
    height = 470
    width = 905
    p = figure(title="Coolant Fluid Temperature ", plot_height=height, plot_width=width)
    p.line(x = x, y = y , color='red', name='g2', line_alpha=1)    
    p.xaxis.major_label_text_font_size = "20pt"
    p.yaxis.major_label_text_font_size = "20pt"
    p.legend.label_text_font_size = '20pt'
    p.title.text_font_size = '28pt'
    p.toolbar.logo = None
    p.toolbar_location = None
    p.xaxis.major_label_orientation = np.pi/4
    p.title.align ="center"
    p.xaxis[0].formatter = PrintfTickFormatter(format="%4.2e")
    p.xaxis.axis_label = 'Time'
    p.yaxis.axis_label = 'Temperature'
    p.xaxis.axis_label_text_font_size = '24pt'
    p.yaxis.axis_label_text_font_size = '24pt'

    
    show(p)
export_png(p, filename="/home/torenvln/Desktop/anomaly_example.png")

# Title: 28
# x-y axes: 24
# x-y values: 20
```
