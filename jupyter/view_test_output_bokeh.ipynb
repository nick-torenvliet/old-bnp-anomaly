```python
# the required python libraries imported
import bnpy
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
from glob import glob
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, push_notebook
from bokeh.core.validation import silence
from bokeh.core.validation.warnings import MISSING_RENDERERS
from bokeh.layouts import column
from IPython.core.display import display, HTML
import bokeh
bokeh.io.reset_output()

bokeh.io.output_notebook()
```

```python
data_dir = "../data/test_output/machine_temp_sys_fail"

ds= [
        "t"   
]

# data_dir = "../data/test_output/bs-5_wsib-5"

# ds= [
#         "ds00",
#         "ds01",
#         "ds02",
#         "ds03",
#         "ds04",
#         "ds05",
#         "ds06",
#         "ds07"    
# ]

p = []
c = []
for i, name in enumerate(ds):
    p.append([])
    c.append([])
    p[i].append(figure(title=name + " DATA" , plot_height=250, plot_width=1200))
    p[i].append(figure(title=name + " log likelihood " + str(i), plot_height=250, plot_width=1200, x_range=p[i][0].x_range))
    p[i].append(figure(title=name + " approx log likelihood " + str(i), plot_height=250, plot_width=1200, x_range=p[i][0].x_range))
    p[i].append(figure(title=name + " entropy", plot_height=250, plot_width=1200, x_range=p[i][0].x_range))
    c[i].append(column(p[i][0], p[i][1], p[i][2], p[i][3]))


for i, name in enumerate(ds):

    sim_string0 = '.*alg-0*.csv'
    sim_string1 = '.*alg-1*.csv'
    sim_string2 = '.*alg-2*.csv'
    g0 =    sorted(glob(data_dir + "/" + name + sim_string0))
    g1 =    sorted(glob(data_dir + "/" + name + sim_string1))
    g2 =    sorted(glob(data_dir + "/" + name + sim_string2))

    df0 = pd.DataFrame()
    df1 = pd.DataFrame()
    df2 = pd.DataFrame()

    for f in g0:
        df0 = df0.append(pd.read_csv(f))
    for f in g1:
        df1 = df1.append(pd.read_csv(f))
    for f in g2:
        df2 = df2.append(pd.read_csv(f))

    df0.set_index('x', inplace=True)   
    df1.set_index('x', inplace=True)   
    df2.set_index('x', inplace=True)   

    calc0 = df0.dropna()
    calc1 = df1.dropna()
    calc2 = df2.dropna()

    p[i][0].line(x = df0.index.tolist(), y = df0['y'].tolist() , color='red', name='g2')
    p[i][1].scatter(x = calc0.index.tolist(), y = calc0['LL'].tolist() , color='red', name='g4')
    p[i][1].scatter(x = calc1.index.tolist(), y = calc1['LL'].tolist() , color='green', name='g4')
    p[i][1].scatter(x = calc2.index.tolist(), y = calc2['LL'].tolist() , color='blue', name='g4')
    p[i][2].scatter(x = calc0.index.tolist(), y = calc0['approx_LL'].tolist() , color='red', name='g4')
    p[i][2].scatter(x = calc1.index.tolist(), y = calc1['approx_LL'].tolist() , color='green', name='g4')
    p[i][2].scatter(x = calc2.index.tolist(), y = calc2['approx_LL'].tolist() , color='blue', name='g4')
    p[i][3].scatter(x = calc0.index.tolist(), y = calc0['entropy'].tolist() , color='red', name='g4')
    p[i][3].scatter(x = calc1.index.tolist(), y = calc1['entropy'].tolist() , color='green', name='g4')
    p[i][3].scatter(x = calc2.index.tolist(), y = calc2['entropy'].tolist() , color='blue', name='g4')

    show(c[i][0])
```
