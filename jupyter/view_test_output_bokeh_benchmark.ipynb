```python
# the required python libraries imported
import bnpy
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
from glob import glob
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, push_notebook
from bokeh.core.validation import silence
from bokeh.core.validation.warnings import MISSING_RENDERERS
from bokeh.layouts import column
from IPython.core.display import display, HTML
import bokeh
from bokeh.models import PrintfTickFormatter
bokeh.io.reset_output()

bokeh.io.output_notebook()
```

```python
data_dir = "../data/test_output/benchmark"

ds= [
        "t"   
]

p = []
c = []
for i, name in enumerate(ds):
    p.append([])
    c.append([])
    height = 470
    width = 905
    p[i].append(figure(title=name[0], plot_height=height, plot_width=width))
    p[i].append(figure(title="Incoming Batch Log Likelihood", plot_height=height, plot_width=width, x_range=p[i][0].x_range))
    p[i].append(figure(title="Incoming Batch Approximate Log Likelihood", plot_height=height, plot_width=width, x_range=p[i][0].x_range))
    p[i].append(figure(title="Entropy Across Window Log Likelihood", plot_height=height, plot_width=width, x_range=p[i][0].x_range))
    p[i].append(figure(title="Entropy Across Window Approximate Log Likelihood", plot_height=height, plot_width=width, x_range=p[i][0].x_range))
    c[i].append(column(p[i][0], p[i][1], p[i][2], p[i][3], p[i][4]))

for i, name in enumerate(ds):

    sim_string0 = '.*alg-0*.csv'
    sim_string1 = '.*alg-0*.csv'
    sim_string2 = '.*alg-0*.csv'
    g0 =    sorted(glob(data_dir + "/" + name + sim_string0))
    g1 =    sorted(glob(data_dir + "/" + name + sim_string1))
    g2 =    sorted(glob(data_dir + "/" + name + sim_string2))

    df0 = pd.DataFrame()
    df1 = pd.DataFrame()
    df2 = pd.DataFrame()

    for f in g0:
        df0 = df0.append(pd.read_csv(f))
    for f in g1:
        df1 = df1.append(pd.read_csv(f))
    for f in g2:
        df2 = df2.append(pd.read_csv(f))

    df0.set_index('x', inplace=True)   
    df1.set_index('x', inplace=True)   
    df2.set_index('x', inplace=True)   

    calc0 = df0.dropna()
    calc1 = df1.dropna()
    calc2 = df2.dropna()


    
    data = name[0]
    g0LL = "SMOVB"
    g0ALL ="SMOVB"
    g0E ="SMOVB"
#    g1LL ="cohesion 1"
#    g1ALL ="cohesion 1"
#    g1E ="cohesion 1"
    g2LL ="TRDPMAD"
    g2ALL ="TRDPMAD"
    g2E ="TRDPMAD"
    
    p[i][0].line(x = df0.index.tolist(), y = df0['y'].tolist() , color='red', name='g2', line_alpha=1, legend_label=data)
    
    p[i][1].line(x = calc0.index.tolist(), y = calc0['ll'].tolist() , color='red', name='g4', line_alpha=0.4, legend_label=g0LL)
#     p[i][1].line(x = calc1.index.tolist(), y = calc1['ll'].tolist() , color='green', name='g4', line_alpha=0.4, legend_label=g1LL)
    p[i][1].line(x = calc2.index.tolist(), y = calc2['ll'].tolist() , color='blue', name='g4', line_alpha=0.4, legend_label=g2LL)
    p[i][2].line(x = calc0.index.tolist(), y = calc0['a_ll'].tolist() , color='red', name='g4', line_alpha=0.4, legend_label=g0ALL)
#     p[i][2].line(x = calc1.index.tolist(), y = calc1['a_ll'].tolist() , color='green', name='g4', line_alpha=0.4, legend_label=g1ALL)
    p[i][2].line(x = calc2.index.tolist(), y = calc2['a_ll'].tolist() , color='blue', name='g4', line_alpha=0.4, legend_label=g2ALL)
    p[i][3].line(x = calc0.index.tolist(), y = calc0['e_ll'].tolist() , color='red', name='g4', line_alpha=0.4, legend_label=g0E)
#     p[i][3].line(x = calc1.index.tolist(), y = calc1['e_ll'].tolist() , color='green', name='g4', line_alpha=0.4, legend_label=g1E)
    p[i][3].line(x = calc2.index.tolist(), y = calc2['e_ll'].tolist() , color='blue', name='g4', line_alpha=0.4, legend_label=g2E)
    p[i][4].line(x = calc0.index.tolist(), y = calc0['e_a_ll'].tolist() , color='red', name='g4', line_alpha=0.4, legend_label=g0E)
#     p[i][4].line(x = calc1.index.tolist(), y = calc1['e_a_ll'].tolist() , color='green', name='g4', line_alpha=0.4, legend_label=g1E)
    p[i][4].line(x = calc2.index.tolist(), y = calc2['e_a_ll'].tolist() , color='blue', name='g4', line_alpha=0.4, legend_label=g2E)
    for j in range(len(p[i])):
        p[i][j].legend.location = "bottom_right"
        p[i][j].xaxis.major_label_text_font_size = "12pt"
        p[i][j].yaxis.major_label_text_font_size = "12pt"
        p[i][j].legend.label_text_font_size = '10pt'
        p[i][j].title.text_font_size = '14pt'
        #p[i][j].toolbar.logo = None
        #p[i][j].toolbar_location = None
        p[i][j].xaxis.major_label_orientation = np.pi/4
        p[i][j].title.align ="center"
        p[i][j].xaxis[0].formatter = PrintfTickFormatter(format="%4.1e")


    show(c[i][0])
```
