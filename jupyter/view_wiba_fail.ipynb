```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt

data_dir = [6]

for d in data_dir:
    All_files = sorted(glob('/home/torenvln/git/bnp-anomaly/data/fail/' + str(d) + '/*.csv'))
    print(All_files)
    headers = ('x','y')
    for file in All_files:
        name = file.split('.csv')[0]
        df = pd.read_csv(file, names=headers)
        x,y = df['x'], df['y']
        fig, ax = plt.subplots(figsize=(7,4))
        ax.plot(x,y)
        plt.tight_layout()
        plt.title(file)
        #plt.savefig(name+'.png', dpi=300, bbox='tight')
        plt.show()
```
